## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
import math
df = pd.read_csv(r"../../data\raw\content_refresh_anonymized.csv")
```

```python
median_impressions = df["impressions_90d"].median()
median_ctr = df["ctr"].median()

candidate_pages = df[
    (df["impressions_90d"] > median_impressions) &
    (df["ctr"] < median_ctr)
]
candidate_pages.shape
```

```python
last_update_median = candidate_pages["days_since_last_update"].median()
print(last_update_median)
```

```python
trend_down = candidate_pages[
    candidate_pages["trend_direction"] == "down"
]
trend_down.shape
percentage_trend_down=(len(trend_down)/len(candidate_pages))*100
print(f"Down trending pages among 3390 cnadidate pages are {math.floor(percentage_trend_down)}%")
```

Interpretation: The analysis identified 3,309 webpages with above median impressions and below median click through rates, indicating that these pages receive substantial visibility but are not converting impressions into clicks effectively. Among these pages, the median time since the last update was 22 days, suggesting that many have been updated recently and that low CTR may not be solely due to outdated content. Additionally, 2,234 of these candidate pages were found to be trending downward, indicating that a significant portion are experiencing declining performance. These findings suggest that such pages should be prioritized for further review, as improving their content, titles, or metadata may help increase user engagement and overall performance.
